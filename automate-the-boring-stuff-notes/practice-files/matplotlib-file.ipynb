```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# sample plot
print(type(plt))
plt.plot([1,2, 3], [2, 4, 6],'b^--', label='2x', color='red', linewidth=2, markersize=10)

# adding a text
plt.title('sample plot',fontdict={'fontname': 'Comic sans MS', 'fontsize': 18})
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot([1, 3, 5, 7, 11], [4, 8, 12, 16, 20])
# changing to floating points
# plt.xticks([0, 1, 3, 5])
# plt.yticks([1, 3, 5, 10])
new_x_axis = np.arange(0, 4.5, 0.5)
print(new_x_axis)
plt.plot(new_x_axis, new_x_axis**2, label="x^2")
# adding a legend
plt.legend() # for the legend to show, we must set the label of the legend


plt.show()
```

```python
x1 = [0, 2, 3, 4, 5]
y1 = [3, 1, 9, 12, 15]

x2 = [0, 2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9, 11]

plt.figure(figsize=(5, 3.5), dpi=300)

plt.plot(x1, y1, 'r', label="2x")
plt.plot(x2[:3], y2[:3], 'b', label="2x")
plt.plot(x2[2:], y2[2:], 'g--', label='2x')
plt.title('First plot')
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.savefig('static/graph.png', dpi=300)

plt.legend()

plt.show()
```
