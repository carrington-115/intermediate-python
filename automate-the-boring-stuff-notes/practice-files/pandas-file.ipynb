```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_excel('static/projections.xlsx')
plt.title('SWIMS 5 Years Financial Projections')
plt.ylabel('Amount in USD')
plt.xlabel('Year')

plt.plot(data['Year'], data['Revenue (USD)'], 'y', label="Revenue")
plt.plot(data['Year'], data['Expenses (USD)'], 'r', label="Expenses")
plt.plot(data['Year'], data['Net Profit (USD)'], 'g--4', label="Net profit")
plt.legend()

plt.savefig('static/projections.png', dpi=200)

plt.show()
```
